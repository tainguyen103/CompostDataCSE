# 2023 Figures

```python
import pandas as pd
import numpy as np
```

```python
# read in the dataset from Excel, name it fy23
fy23 = pd.read_excel(io="/Users/tainguyen103/Downloads/MASTER Waste 2011-2023.xlsx", 
                     sheet_name="FY23Compost",
                     header=1)
```

```python
# display the first 5 rows of dataset
fy23.head()
```

```python
# display the last 5 rows of dataset. There has been some removals on some last lines on the local excel file to help reflect null values more accurately. 
fy23.tail()
```

```python
# count the number of the "Weekly Totals" values in the Date column
week_counts23 = fy23["Date"].value_counts()["Weekly Totals"]
week_counts23
```

### There are 48 weeks during FY23. We are going to multiple this number by 21 since each week has 7 days and there are 3 meals served each day. Notice that there are two special weeks : first one served only for 2 days and last one served for 6 days. So we will subtract 5 + 1 = 6 days (or 18 meals) from the total

```python
# count the total meals of 2023
total_meals23 = 48 * 21 - 6 * 3
print("Total meals in FY23: ",total_meals23)
```

```python
# count the number of null values for breakfast, subtract it by 48 since we do not count the null values of "Weekly Totals"
null_bf23 = fy23["Breakfast"].isnull().value_counts()[True] - 48 
```

```python
# count the number of null values for lunch, subtract it by 48 since we do not count the null values of "Weekly Totals"
null_ln23 = fy23["Lunch"].isnull().value_counts()[True] - 48
```

```python
# count the number null values for dinner, subtract it by 48 since we do not count the null values of "Weekly Totals"
null_dn23 = fy23["Dinner"].isnull().value_counts()[True] - 48
```

```python
# print out total number of null values for all 3 meals
print(null_bf23 + null_ln23 + null_dn23)
```

```python
# calculate the percentage of null meals in total meals
perc23 = (434/990)*100
perc23
```

```python
# count the number of green buckets
total_gb23 = fy23["Daily Totals"].iloc[:378].sum()
total_gb23
```

```python
# count the number of days on which meals are served
total_days23 = 48 * 7 - 6 
total_days23
```

```python
# calculate the average number of green buckets for each day
average_gb23 = total_gb23 / total_days23
average_gb23
```

```python
# filter by dates that used below the average number of green bucket, which is less than or equal to 5.68
fy23['Daily Totals'] = fy23['Daily Totals'].astype(str)
daycount23_belowAVG = fy23.loc[fy23['Daily Totals'].isin(['1','2', '3', '4', '5'])]
daycount23_belowAVG
```

### According to the filtered table we can see that there are 117 rows (or dates) with the number of green buckets used falling below the average. 

```python
# filter by dates where all 3 meals are null or do not have a recorded value
nullallmeals23 = fy23.loc[(fy23['Breakfast'].isnull()) & (fy23['Lunch'].isnull()) & (fy23['Dinner'].isnull())]
actual23 = nullallmeals23.loc[nullallmeals23["Date"] != "Weekly Totals"]
actual23
```

```python
# count the days, or number of rows
actual23.shape[0]
```
